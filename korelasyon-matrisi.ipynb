```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Veri setini ykleyelim (varsaylan olarak "diabetes.csv" olarak kabul edelim)
data = pd.read_csv('/kaggle/input/diabetes-data-set/diabetes.csv')

# zellikler arasndaki korelasyon matrisini hesaplayalm
correlation_matrix = data.corr()

# Korelasyon analizi grafiini izelim
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, square=True, cbar_kws={"shrink": 0.8, "aspect": 30})
plt.title("zellikler Arasndaki Korelasyon Matrisi", fontsize=18, fontweight='bold', pad=20)
plt.xticks(fontsize=12, rotation=45, ha='right')
plt.yticks(fontsize=12)
plt.xlabel('zellikler', fontsize=14, labelpad=10)
plt.ylabel('zellikler', fontsize=14, labelpad=10)
plt.tight_layout()
plt.show()



```
